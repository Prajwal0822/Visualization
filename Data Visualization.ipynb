```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition 
from sklearn import datasets 
```

```python
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
d = pd.Series(iris)
```

```python
d['target_names']
```

```python
from sklearn.decomposition import PCA
X = PCA(n_components=3).fit_transform(iris.data)#Dimensionality Reduction Using PCA for iris data
```

```python
Y = iris.target
```

```python
X
```

```python
Y
```

```python
fig = plt.figure(1,figsize=(8,6))
ax = Axes3D(fig,elev=-150,azim=110)
ax.scatter(X[:,0],X[:,1],X[:,2],c=Y,s=100,cmap=plt.cm.rainbow_r)
ax.set_title('Data Visualisation')
ax.set_xlabel('PC1',labelpad=15)
ax.set_ylabel('PC2',labelpad=15)
ax.set_zlabel('PC3',labelpad=15)
plt.show()
```
